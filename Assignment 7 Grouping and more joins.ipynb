# Assignment 7: Grouping and more joins

#### Exercise / Query 1
Using a single join and a grouping, write a query that allows you to find the forename, surname, driverid and the total number of wins for each driver in the f1db schema.

```python
%%sql  select q.forename, q.surname, q.driverid, foo.p AS wins from f1db.drivers AS q 
INNER JOIN (select driverid, count(positionorder) AS p from f1db.results WHERE positionorder=1 group by driverid) AS foo 
ON q.driverid=foo.driverid GROUP BY q.driverid, foo.p ORDER BY foo.p DESC;
```

#### Exercise / Query 2
Write a query that joins the tables results, constructors and drivers to show the amount of times a driver has driven a car from a constructor. In other words how many times one driver (for instance Schumacher) has driven a constructor (for instance Mercedes). Your table should have three fields: drivers.driverref, constructors.name and count(*) (showing the number of times the driver has driven the constructor).

```python
%%sql select count(r.constructorid) AS times_driven, d.driverref, c.constructorref from f1db.results AS r 
INNER JOIN (select driverid, driverref from f1db.drivers) AS d 
ON r.driverid=d.driverid INNER JOIN (select constructorid, constructorref from f1db.constructors) AS c 
ON r.constructorid=c.constructorid 
GROUP BY c.constructorref, d.driverref 
ORDER BY count(r.constructorid) 
DESC;
```

#### Exercise / Query 3
Now we have found the most driven vehicles even for those who didn't finish the race. Extend the query from 2 by removing all tuples from the result where the status is not "Finished".

```python
%%sql select count(r.constructorid) AS times_driven, d.driverref, c.constructorref from f1db.results AS r 
INNER JOIN (select driverid, driverref from f1db.drivers) AS d 
ON r.driverid=d.driverid INNER JOIN (select constructorid, constructorref from f1db.constructors) AS c 
ON r.constructorid=c.constructorid 
WHERE r.statusid = 1
GROUP BY c.constructorref, d.driverref 
ORDER BY count(r.constructorid) 
DESC;
```

#### Exercise / Query 4
Create a third (and new) query that finds the amount of milliseconds spent in pitstop (see the pitstops relation) by each unique combinations of driverid and raceid.

```python
%%sql CREATE VIEW totaltime_pitstops AS select d.driverref, s.raceid, sum(s.milliseconds) from f1db.pitstops AS s 
INNER JOIN (select driverid, driverref from f1db.drivers) AS d 
ON s.driverid=d.driverid 
GROUP BY d.driverref, s.raceid 
ORDER BY sum(s.milliseconds) DESC;
```

#### Exercise / Query 5
Time to put everything together. Use the query for 4 as a join subquery in query 3 (see slides on "a query within a query") to find the total amount of pitstop time, for each result in query 3

```python
%%sql select count(r.constructorid) AS times_driven, d.driverref, c.constructorref, sum(totalmil.ms) from f1db.results AS r 
INNER JOIN (select driverid, driverref from f1db.drivers) AS d 
ON r.driverid=d.driverid 
INNER JOIN (select constructorid, constructorref from f1db.constructors) AS c 
ON r.constructorid=c.constructorid 
INNER JOIN (select driverref, sum(s.milliseconds) AS ms from f1db.pitstops AS s
INNER JOIN (select driverid, driverref from f1db.drivers) as innerdriver
ON s.driverid=innerdriver.driverid GROUP BY innerdriver.driverref) AS totalmil
ON totalmil.driverref=d.driverref
WHERE r.statusid=1  
GROUP BY c.constructorref, d.driverref
ORDER BY count(r.constructorid) DESC;
```

#### Exercise / Query 6
As a last thing we want to find the driver with the least pistop time. However we want to control for the amount of races (see explanation below), so we need to include one last column: the average pitstop time per race. Your table should be sorted based on the ascending values of that average pitstop time.  
  
**Explanation**: If we just count the pitstop time, we only get the total time he spend in a workshop, no matter how many races he participated in. If a driver participated in 100 races, he would have more pitstop time than a driver participating in one race!

```python
%%sql select count(r.constructorid) AS times_driven, d.driverref, c.constructorref, sum(totalmil.ms), avg(totalmil.ms) from f1db.results AS r 
INNER JOIN (select driverid, driverref from f1db.drivers) AS d 
ON r.driverid=d.driverid 
INNER JOIN (select constructorid, constructorref from f1db.constructors) AS c 
ON r.constructorid=c.constructorid 
INNER JOIN (select driverref, sum(s.milliseconds) AS ms from f1db.pitstops AS s
INNER JOIN (select driverid, driverref from f1db.drivers) as innerdriver
ON s.driverid=innerdriver.driverid GROUP BY innerdriver.driverref) AS totalmil
ON totalmil.driverref=d.driverref
WHERE r.statusid=1  
GROUP BY c.constructorref, d.driverref
ORDER BY avg(totalmil.ms) ASC LIMIT 1;
```
